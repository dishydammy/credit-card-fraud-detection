```python
from sklearn.model_selection import train_test_split
```

```python
df_full_train, df_test = train_test_split(df, test_size=0.2, stratify=df['class'], random_state=10)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, stratify=df_full_train['class'], random_state=10)
```

```python
print(df_full_train['class'].value_counts(normalize=True))
print(df_train['class'].value_counts(normalize=True))
print(df_val['class'].value_counts(normalize=True))
print(df_test['class'].value_counts(normalize=True))
```

```python
y_train = df_train['class']
y_val = df_val['class']
y_test = df_test['class']
```

```python
del df_train['class']
del df_val['class']
del df_test['class']
```

```python
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(max_iter=10000, class_weight='balanced', random_state=10)
log_model.fit(df_train, y_train)
```

```python
from sklearn.metrics import (
    classification_report, confusion_matrix, 
    roc_curve, auc, precision_recall_curve, average_precision_score
)
```

```python
y_pred = log_model.predict(df_val)
y_pred_proba = log_model.predict_proba(df_val)[:, 1]
```

```python
print("Classification Report:\n", classification_report(y_val, y_pred))
```

```python
cm = confusion_matrix(y_val, y_pred)
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=[0,1], yticklabels=[0,1])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```

```python
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.plot([0,1], [0,1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
precision, recall, thresholds = precision_recall_curve(y_val, y_pred_proba)
avg_precision = average_precision_score(y_val, y_pred_proba)

plt.figure(figsize=(6,6))
plt.plot(recall, precision, label=f"PR Curve (AP = {avg_precision:.2f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.show()
```

```python
def predict(model, df_train, y_train):
    model.fit(df_train, y_train)

    y_pred = model.predict(df_val)
    y_pred_proba = model.predict_proba(df_val)[:, 1]

    print("Classification Report:\n", classification_report(y_val, y_pred))
    print("------------------------------------------------------------")
    print('\n')
    
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    cm = confusion_matrix(y_val, y_pred)
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=axes[0],
                xticklabels=model.classes_, yticklabels=model.classes_)
    axes[0].set_title("Confusion Matrix")
    axes[0].set_xlabel("Predicted")

    print("------------------------------------------------------------")
    print('\n')
    
    fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    axes[1].plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
    axes[1].plot([0,1], [0,1], linestyle="--", color="gray")
    axes[1].set_title("ROC Curve")
    axes[1].set_xlabel("False Positive Rate")
    axes[1].set_ylabel("True Positive Rate")
    axes[1].legend()

    print("------------------------------------------------------------")
    print('\n')

    precision, recall, _ = precision_recall_curve(y_val, y_pred_proba)
    avg_precision = average_precision_score(y_val, y_pred_proba)
    axes[2].plot(recall, precision, label=f"AP = {avg_precision:.2f}")
    axes[2].set_title("Precision-Recall Curve")
    axes[2].set_xlabel("Recall")
    axes[2].set_ylabel("Precision")
    axes[2].legend()
    
    plt.tight_layout()
    plt.show()  
```

```python
xgb_model = xgb.XGBClassifier(
    objective="binary:logistic",
    scale_pos_weight=(y_train.value_counts()[0] / y_train.value_counts()[1]),
    eval_metric="logloss",
    use_label_encoder=False,
    random_state=42)
```

```python
xgb_param_dist = {
    "n_estimators": [100, 200, 300],
    "max_depth": [3, 5, 7],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.7, 0.8, 1.0],
    "colsample_bytree": [0.7, 0.8, 1.0]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

xgb_search = RandomizedSearchCV(
    estimator=xgb_model,
    param_distributions=xgb_param_dist,
    n_iter=10, 
    scoring="f1",
    cv=5,
    verbose=1,
    random_state=42,
    n_jobs=-1
)
```

```python
xgb_search.fit(df_train, y_train)
xgb_best = xgb_search.best_estimator_
print("Best XGB params:", xgb_search.best_params_)
```

```python
predict(xgb_best, df_train, y_train)
```

```python
predict(xgb_best, X_train_res, y_train_res)
```

```python
lgb_model = lgb.LGBMClassifier(
    objective="binary",
    class_weight="balanced",
    random_state=10)
```

```python
lgb_param_dist = {
    "n_estimators": [100, 200, 300],
    "max_depth": [-1, 5, 7],
    "learning_rate": [0.01, 0.05, 0.1],
    "num_leaves": [15, 31, 63],
    "subsample": [0.7, 0.8, 1.0]
}
```

```python
lgb_search = RandomizedSearchCV(
    estimator=lgb_model,
    param_distributions=lgb_param_dist,
    n_iter=10, 
    scoring="f1",
    cv=5,
    verbose=1,
    random_state=10,
    n_jobs=-1
)
```

```python
lgb_search.fit(df_train, y_train)
lgb_best = lgb_search.best_estimator_
print("Best LGB params:", lgb_search.best_params_)
```

```python
predict(lgb_best, df_train, y_train)
```

```python
predict(lgb_best, X_train_res, y_train_res)
```
